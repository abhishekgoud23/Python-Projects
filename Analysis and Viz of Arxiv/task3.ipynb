### Importing required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing csv file as data frame. Also creating data frames of each category.

```python
df = pd.read_csv("task1.csv")                          #Reading csv file using read_csv() of pandas library to data frame
dfAR = df.loc[(df['primary_category'] == 'cs.AR')]     #dfAR is dataframe whose primary category is cs.AR
dfLG = df.loc[(df['primary_category'] == 'cs.LG')]     #dfLG is dataframe whose primary category is cs.LG
dfCC = df.loc[(df['primary_category'] == 'cs.CC')]     #dfCC is dataframe whose primary category is cs.CC
dfAI = df.loc[(df['primary_category'] == 'cs.AI')]     #dfAI is dataframe whose primary category is cs.AI
```

### Below Function is used to create dictionary with keys as 1,2,3 or 4,>4 authors and values with count of each.

```python
def createDict(ls):
    dic = {}                              #Temporary dictionary
    for i in range(len(ls)):              #Traversing completely through the list
        eleLen = ls[i].count(',')+1       #Counts number of seperations and adding 1 to that sum gives num of authors in that list
        if eleLen==1:                     #If element list is 1
            k = 'One Author'              #setting key as 'One Authors'
        elif eleLen==2:                   #If element list is 2
            k = 'Two Authors'             #setting key as 'Two Authors'
        elif eleLen==3 or eleLen==4:      #If element list is 3 or 4
            k = '3 or 4 Authors'          #setting key as '3 or 4 Authors' 
        else:                             #If element list is >4
            k = 'More than 4 Authors'     #setting key as 'More than 4 Authors'
        if k in dic:                      #Filling dictionary
            dic[k] += 1
        else: 
            dic[k] = 1
    return (dic)                          #returns populated dictionary
```

### Extracting authors from each category dataframe and passing to above function to create dictionaries for each category

```python
lsAI = dfAI["authors"].tolist() #Extracting authors from dataframe and store it to list
lsCC = dfCC["authors"].tolist() #Extracting authors from dataframe and store it to list
lsLG = dfLG["authors"].tolist() #Extracting authors from dataframe and store it to list
lsAR = dfAR["authors"].tolist() #Extracting authors from dataframe and store it to list

dicAI = createDict(lsAI)        #Using created list to create dictionary
dicCC = createDict(lsCC)        #Using created list to create dictionary
dicLG = createDict(lsLG)        #Using created list to create dictionary
dicAR = createDict(lsAR)        #Using created list to create dictionary
print(dicAI)
print(dicCC)
print(dicLG)
print(dicAR)
```

### For each category, plotting pie chart for given factors

```python
numOfAuthorsAI = []                   #Stores the key value
valAI = []                            #Stores number of values 

for x, y in dicAI.items():            
    numOfAuthorsAI.append(x)          #Appending keys to numOfAuthorsAI
    valAI.append(y)                   #Appending values to valAI

plt.pie(valAI, labels=numOfAuthorsAI,autopct='%1.1f%%') #Plots pie value of valAI with labels and percentage values.
plt.axis('equal')                     #Pie will always be rendered as a circle because of the equal aspect ratio.
plt.show()                            #To display plot
```
