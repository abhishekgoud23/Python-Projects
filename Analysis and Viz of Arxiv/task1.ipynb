### Importing required libraries

```python
import arxiv              #Importing arxiv library 
import time
import csv
import pandas as pd
import numpy as np
```

#### Below Function is used to createIDs 

```python
ans = []                                #Stores list of lists of year snd month wise id
def createIDs(yr,mo):                   #year and months as parameter
    ids = []                            #Temporary list that collects ids created
    i = '00001'                         #Variable that iterates file id
    for l in range(1,100000):           #Iterating from 1 to 99999
        s = yr + mo.zfill(2)+'.'+str(l).zfill(5) #Creating id in string format
        ids.append(s)                   #Appending to ids list
    return ids                          #Returning ids list of a year and a month
```

#### Below block of code is used to create ids and store in "ans" list of given years(2017 to 2021)

```python
yr = [i for i in range(17,22)]      #Creating list of years(17 to 21)
mo = [i for i in range(1,13)]       #Creating list of years(1 to 12)
print(yr,mo)
for i in yr:                        #Iterating through each year
    for j in mo:                    #Iterating through each month of ith year
        ans.append(createIDs(str(i),str(j))) #appending output of createIDs function
```

#### Below code is used to collect metadata of a given ids

I am breaking big list into small chunks as per the idea in arxiv_helper.ipynb file

```python
#Reference arxiv_helper.ipynb file

res = []                                  #This is list of dictionaries of metadata collected
def docsDet(ids):                         #Parameter is list of ids
    st = 0                                #Lower limit
    end = 500                             #Upper limit
    break_flag = False                        
    cnt = 0                               #Counter for cummulative results
    while st != 100000:                   #Breaking loop as soon as lower limit reaches 100000
        print("in")
        search = arxiv.Search(id_list=ids[st:end]) #Search function of arxiv library accepts list of ids to search and get metadata
        result= list(search.results())    #
        print("done")

        if result == []:
            break_flag = True #Set flag that loop was broken out of
            break
        cntType = {'cs.LG':0,'cs.AI':0,'cs.CC':0,'cs.AR':0}

        for i in range(len(result)):
            paper_info = {}
            if result[i].primary_category in ['cs.LG','cs.AI','cs.CC','cs.AR']:
                cnt += 1 #Increment counter
                cntType[result[i].primary_category] += 1 #Increment specific ctr
                paper_info['title'] = result[i].title
                paper_info['authors'] = []
                for author in result[i].authors:
                    paper_info['authors'].append(str(author))
                paper_info['summary'] = result[i].summary
                paper_info['comment'] = result[i].comment
                paper_info['journal_ref'] = result[i].journal_ref
                paper_info['doi'] = result[i].doi
                paper_info['entry_id'] = result[i].entry_id
                paper_info['updated'] = str(result[i].updated)
                paper_info['published'] = str(result[i].published)
                paper_info['primary_category'] = result[i].primary_category
                paper_info['categories'] = result[i].categories
                paper_info['links'] =  str(result[i].links)
                paper_info['pdf_url'] = result[i].pdf_url
                res.append(paper_info)

        print("Range:",(st,end))
        print("Cumulative count:",cnt)
        print("Count by type",cntType)
        print("-----")
        st = end #Old high becomes new lo
        end += 500 #New high incremented by 500
        time.sleep(1.5)
```

```python
for p in range(len(ans)):
    docsDet(ans[p])
```
